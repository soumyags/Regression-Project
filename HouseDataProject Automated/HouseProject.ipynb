Automate Preprocessing

```python
from sklearn.pipeline import Pipeline #Automating Preprocessing
from sklearn.impute import SimpleImputer # Replacing Missing Values
from sklearn.preprocessing import StandardScaler,OneHotEncoder # for Categorical & Continuous data(X) preprocessing
from sklearn.compose import ColumnTransformer # to combine numerical and text
```

As our training_set.csv dont have any missing value, but has "NA" values which is not Miising but as python condider it as Missing, so we need to inform python to replace all "NA" values with constant "Not Available" The below steps will replace 

Here we will use SimpleImputer for replacing "NA" values for Categorical feature

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value="Not Available")),
                         ('scaler',StandardScaler())])
```

The below is for the text data to replace "NA" with "Not Available"

```python
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value="Not Available")),
                         ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```
