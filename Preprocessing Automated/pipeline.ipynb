Step3:Seperate X & Y

```python
X=df.drop(columns=['id','Weight'])
Y=df[['Weight']]
```

Seperate Categorical & Continuous data of X

```python
cat=list(X.columns[X.dtypes=='object'])
con=list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline #Automating Preprocessing
from sklearn.impute import SimpleImputer # Replacing Missing Values
from sklearn.preprocessing import StandardScaler,OneHotEncoder # for Categorical & Continuous data(X) preprocessing
from sklearn.compose import ColumnTransformer # to combine numerical and text
```

This below step will remove the missing value of numeric data with 'Mean' & also do the standard scaler for the numeric data i.e. make all the numberic daa in range -3 to 3

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

the below steps wil remove the missing value of Text Data i.e. Airbags columns, using 'Mode' and apply One hot encoding for the same

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

Step 5 : Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2,
                                                random_state=21)
```

Step 6 : Build the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

Evaluate R2 scores in train and test

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

Get MSE, RMSE, MAE, R2

```python
def evaluate_model(model, x, y):
    # Predict the results for x
    ypred = model.predict(x)
    # Import packages
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print mse, rmse, mae, r2
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
